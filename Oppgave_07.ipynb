# Oppgave 7
Lag en liste med tre tall. Undersøk om bare det første tallet er partall, om bare det andre tallet er partall, om bare det tredje tallet er partall, om det er flere partall, eller om det er ingen partall. Skriv ut resultatet.

```python
# Liste med tre tall
liste = [1, 1, 1]

# Gjennomfører undersøkelsene
if liste[0] % 2 == 0 and liste[1] % 2 != 0 and liste[2] % 2 != 0:
    print('Bare det 1. tallet er partall.')
elif liste[0] % 2 != 0 and liste[1] % 2 == 0 and liste[2] % 2 != 0:
    print('Bare det 2. tallet er partall.')
elif liste[0] % 2 != 0 and liste[1] % 2 != 0 and liste[2] % 2 == 0:
    print('Bare det 3. tallet er partall.')
elif liste[0] % 2 == 0 or liste[1] % 2 == 0 or liste[2] % 2 == 0:
    print('Det er flere partall.')
else:
    print('Det er ingen partall.')
```

```python
# Kult alternativ
liste = [1, 1, 1]
partall = [indeks for indeks, tall in enumerate(liste) if tall % 2 == 0]
if len(partall) == 1:
    print(f'Bare det {partall[0]+1}. tallet er partall.')
elif len(partall) > 1:
    print('Det er flere partall.')
else:
    print('Det er ingen partall.')
```
