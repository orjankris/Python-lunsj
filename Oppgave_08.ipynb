# Oppgave 8
Lag en endelig liste med årlige leieinntekter, første inntekt om ett år. Bruk en (diskonterings)rente på 10 %, og beregn nåverdien av denne kontantstrømmen. Hint: Bruk enumerate().

```python
# Inndata
inntekter = [50000, 52000, 54000, 54000, 60000, 60000, 60000]
rente = 0.10
naaverdi = 0

# Beregner nåverdi av kontantstrømmen
for n, inntekt in enumerate(inntekter):
    naaverdi += (inntekt) / (1 + rente) ** (n + 1)

print(f'Nåverdien av leieinntektene er {naaverdi:.0f} kr.')
```

```python
# En mer kompakt variant
inntekter = [50000, 52000, 54000, 54000, 60000, 60000, 60000]
rente = 0.10

naaverdi = sum(inntekt / (1 + rente) ** (n + 1) for n, inntekt in enumerate(inntekter))

print(f'Nåverdien av leieinntektene er {naaverdi:.0f} kr.')
```
